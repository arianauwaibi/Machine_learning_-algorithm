```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm,datasets 
from sklearn.model_selection import train_test_split 
from sklearn import metrics
from sklearn.metrics import plot_confusion_matrix
```

```python
#import dataset 

iris= datasets.load_iris()

x = iris.data[:,2:4]

#Targets are labels 
y= iris.target
```

```python
#Create/Define the Model  
#SVM: Super Vector Machine

C = 1.0 #SVM regularization parameter 

#Classification (the model we trained)
svc = svm.SVC(kernel= "linear", C=C, gamma= "auto").fit(x,y)

#function needed to train the model before testing with the .fit(data,target)
```

```python
#Create mesh to plot the samples

x_min, x_max = x[:,0].min() - 0.5, x[:,0].max() + 0.5

y_min, y_max = x[:,1].min() - 0.5, x[:,1].max() + 0.5

h = (x_max/x_min)/100

xx, yy = np.meshgrid(np.arange(x_min,x_max,h), np.arange(y_min, y_max, h))
```

```python
plt. subplot(111) 

#
z= svc.predict(np.c_[xx.ravel(),yy.ravel()])
z = z.reshape(xx.shape)

plt. contour(xx,yy,z, cmp= plt.cm.Paired,alpha=0.8)
# contour will help it plt the boundaries 

plt.scatter(x[:,0],x[:,1],c=y, cmap = plt.cm.Paired)

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.xlim(xx.min(),xx.max())
plt.title('SVC with Linear kernel')
plt.show()
```
